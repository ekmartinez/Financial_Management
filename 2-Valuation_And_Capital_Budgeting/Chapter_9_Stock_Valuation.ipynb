**1. Stock Values**

The Starr Co. just paid a dividend of 1.90 per share on its stock.  The dividends are expected to grow at a constant rate of 5 percent per year, indefinitely.  If investors require a 12 percent return on the stock, 

**a)** What is the current price?

**b)** What will the price be in three years? 

**c)** In 15 years?

```python
div = 1.90
g = .05
r = .12

a = div * (1 + g)/(r - g)
b = div * (1 + g)**4/(r - g)
c = div * (1 + g)**16/(r - g)

print(f'Current price: {a:.2f}')
print(f'In 3 years: {b:.2f}')
print(f'In 15 years : {c:.2f}')
```

**2. Stock Values**

The next dividend payment by ECY, Inc., will be 2.85 per share.  The dividends are anticipated to maintain a 6 percent growth rate, forever.  If ECY stock currently sells for 58 per share, **what is the required return?**

```python
div = 2.85
g = .06
p = 58

required_return = (div / p) + g
f'Required Rate of Return: {required_return*100:.2f}%'
```

**3. Stock Values**

For the company in the previous problem, what is the dividend yield? What is the expected capital gains yield?

```python
div_yield = div / p
cap_gain_yield = g
print(f'Dividend Yield: {div_yield*100:.2f}%')
print(f'Capital Gains Yield: {g*100:.2f}%')
```

**4. Stock Values**

White Wedding Corporation will pay a 3.05 per share dividend next year.  The company pledges to increase its dividend by 5.25 percent per year, indefinitely.  If you require an 11 percent return on your investment, how much will you pay for the company's stock today?

```python
div = 3.05
g = .0525
r = .11

price_today = div / (r - g)
f'I would pay today: ${price_today:.2f}'
```

**6. Stock Valuation**

Suppose you know that a company's stock currently sells for 64 per share and the required return on the stock is 13 percent. You also know that the total return on the stock is evenly divided between a capital gains yield and a dividend yield.  If it's the company's policy to always maintain a constant growth rate in its dividends, **what is the current dividend per share?**

```python
price = 64
required = .13
div_y = required * .5 # Evenly Divided
div_per_share = price * div_y
print(f'The Current dividend per share is: {div_per_share}')
```

**7. Stock Valuation**

Gruber Corp. pays a constant $11 dividend on its stock. The company will maintain this dividend for the next nine years and will then cease paying dividends forever.  If the required return on this stock is 10 percent, **what is the current share price?**

```python
div = 11
r = .10
prds = 9
price = -npf.pv(fv=0, rate=r, nper=prds, pmt=div, when='end')
print(f'The current share price is: ${price:.2f}')
```

**8. Valuing Preferred Stock**

Ayden Inc., has an issue of preferred stock outstanding that pays a 6.40 dividend every year, in perpetuity.  If this issue currently sells for 103 per share, **what is the required return?**

```python
div = 6.40
price = 103
r = div / price
print(f'The required return is: {r*100:.2f}%')
```

**9. Growth Rate**

The newspape reported last week that Bennington Enterprises earned $28 million this year.  The report also stated that the firm's return on equity is 15 percent.  Bennington retains 70 percent of its earnings.  
1. What is the firm's earnings growth rate?
2. What will next year's earnings be?


```python
earnings = 28 # Millions
roe = .15
retention_ratio = .70

growth_rate = retention_ratio * roe
next_year = earnings * (1 + growth_rate)

print(f'The firm\'s earnings growth rate is: {growth_rate*100:.2f}%')
print(f'Next year\'s earnings will be: ${next_year:.2f} Millions')
```
