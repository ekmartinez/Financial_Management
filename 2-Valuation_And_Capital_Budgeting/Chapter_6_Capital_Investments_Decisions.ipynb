**1. Calculating Project NPV:** Raphael Restaurant is considering the purchase of a $12,000 souffle maker.  The souffle maker has an economic life of five years and will be fully depreciated by the straight-line method.  The machine will produce 1,900 souffles per year, with each costing $2.20 to make and priced at $5.00.  Assume that the discount rate is 14% and the tax rate is 34%.  **Should Raphael make the purchase?**   

```python
# Variable Declaration
equipment_cost = 12000
economic_life = 5
cost_per_unit = 2.20
price_per_unit = 5.00
discount_rate = .14
tax_rate = .34
units_per_year = 1900

# We will solve this by using the Tax Shield Method
sales = units_per_year * price_per_unit
cost = units_per_year * cost_per_unit
depreciation = equipment_cost / economic_life

operating_cash_flows = (sales - cost) * (1 - tax_rate) + depreciation

ocf = [operating_cash_flows for x in range(economic_life)]

npv = -equipment_cost + npf.npv(discount_rate, ocf)
print(f'{npv:,.2f}')
```

```python
import numpy_financial as npf

# Variable Declaration
equipment_cost = 12000
economic_life = 5
cost_per_unit = 2.20
price_per_unit = 5.00
discount_rate = .14
tax_rate = .34
units_per_year = 1900

# We will solve this by using the Tax Shield Method
sales = units_per_year * price_per_unit
cost = units_per_year * cost_per_unit
depreciation = equipment_cost / economic_life

# Correct operating cash flows calculation
operating_cash_flows = (sales - cost) * (1 - tax_rate) + depreciation

# Create a list of cash flows for each year
ocf = [operating_cash_flows for _ in range(economic_life)]

# Calculate NPV
npv = npf.npv(discount_rate, ocf) - equipment_cost
print(f'{npv:,.2f}')
```
