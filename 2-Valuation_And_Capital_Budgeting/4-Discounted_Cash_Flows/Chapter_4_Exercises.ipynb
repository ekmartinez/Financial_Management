## Exercise 16 - Calculating APR

Find the APR, or stated rate, in each of the following cases:

```python
"""
   Effective 
Annual Rate(EAR)     Compounded
----------------------------------
    10.3 %           Semiannually
     9.4               Monthly
     7.2                Weekly
    15.9               Infinite
----------------------------------
"""
```

Lets first define a function that would allow us to convert from EAR to APR.  As you may recall, we find the EAR as follow:

```python
"""
Effective Annual Rate (EAR) = ( 1 + r / m ) ^ m - 1
"""
```

We rearrange the equation to solve for r (APR). 

```python
def aprRate(ear, m):
    apr = m * ((1 + ear)**(1/m) - 1)
    return apr
```

Ok, perfect, now we create our table as shown below.

```python
df = pd.DataFrame([[0.103, 2],
                  [0.094, 12],
                  [0.072, 52]],
                 columns=("EAR", "Compounding"))

df["APR"] = df.apply(lambda row: aprRate(row["EAR"], row["Compounding"]), axis=1)

#Format DataFrame Style.
df['EAR'] = df['EAR'].map('{:.2%}'.format)
df['APR'] = df['APR'].map('{:.2%}'.format)
df = df.style.set_properties(**{'text-align': 'center'})
df
```

Solving APR indefinately is a bit different, since we need to use the `Continuous Compounding` formula.

EAR = e^q - 1

Now we rearrange and solve for APR:

```python
ear = .1590
apr = np.log(1 + ear) * 100
print(f"{apr:.2f}%")
```

## 21. Future Value 

What is the future value in `seven years` of `1,000` invested in an account with a stated annual interest rate of `8 percent`.

1. Compounded annually?
2. Compounded semiannually?
3. Compounded monthly?
4. Compounded continuously?

```python
# Get Variables
n = 7
r = .08
e = 2.718
pmt = 0
pv = -1000
fv = []

# Create a basic data structure
struct = {"Compoundings" : ["Yearly", "Semiannual", "Monthly", "Continuous"],
          "Future_Values" : [] 
         }

# Compoundings
# 1 = Yearly, 2 = Semiannual, 12 = Monthly, e = continuous
compoundings = [1, 2, 12, e]

# Iterate over the compounding list and compute future values.
# Also populate Future_Values in struct.

for m in compoundings:
    if m == e:
        fvlue = np.abs(pv) * (e**(r*n))
        struct["Future_Values"].append(fvlue)
    else:
        comp_rate = r / m
        comp_perd = n * m
        fvlue = npf.fv(pv=pv, rate=comp_rate, nper=comp_perd, pmt=0)
        struct["Future_Values"].append(fvlue)

# Create and Display answer table
df = pd.DataFrame(struct)
df['Future_Values'] = df['Future_Values'].map('{:,.2f}'.format)
display(HTML(df.to_html(index=False)))
```

## 23. Calculating Annuities

You are planning to save for retirement over the next 30 years.  To do this, you will invest 700 dollars a month in a stock account and 300 a month in a bond account.  The return of the stock account is expected to be 10 percent, and the bond account will pay 6 percent.  When you retire, you will combine your money into an account with an 8 percent return. How much can you withdraw each month from your account assumming a 25 year withdrawal period?

Since the investing period ends at the same time as the withdrawals begins, the present value of the retirement payments is equal to the future value of the retirement savings. So we need to get the FV of the stock account and the fv of the bond account and add them together.

```python
save_period = 30
stock_ret = .10
bond_ret = .06
cmp = 12 # Assuming monthly compounding
stock_pmts = 700
bond_pmts = 300

struct = {
    "Stock_Account" : -npf.fv(pv=0, nper=save_period*cmp, rate=stock_ret/cmp, pmt=stock_pmts),
    "Bond_Account" : -npf.fv(pv=0, nper=save_period*cmp, rate=bond_ret/cmp, pmt=bond_pmts)
    }

savings = pd.DataFrame(struct, index=["Future Value"])
savings["Total_Savings"] = savings["Stock_Account"] + savings["Bond_Account"]
savings["Stock_Account"] = savings["Stock_Account"].map('{:,.2f}'.format)
savings["Bond_Account"] = savings["Bond_Account"].map('{:,.2f}'.format)
savings["Total_Savings"] = savings["Total_Savings"].map('{:,.2f}'.format)
savings
```

After the end of our saving period we have accumulated `1,883,696 dollars` which are ready for funding our retirement.  We find the periodic distribution by using the `pmt` function.

```python
r = .08 # combined funds at an 8 percent account
n = 25 # Estimated Withdrawal period
cmp = 12 # Assuming monthly compounding
pv = struct["Stock_Account"] + struct["Bond_Account"]
distribution = -npf.pmt(rate=r/cmp, nper=n*cmp, pv=pv, fv=0, when='end')
print(f"Expected Monthly Distribution: ${distribution:,.2f}")
```

## 26. Growing Perpetuities

Mark Weinstein has been working on an advanced technology in laser eye surgery.  His technology will be available in the near term.  He anticipates his first annual cash flow from the technology to be `215,000`, received `two years from today`.  Subsequent cash flows will `grow` at `4 percent` in `perpetuity`.  What is the `present value` of the technology if the discount rate is `10 percent`?

To solve this problem, we first need to get the present value of the growing perpetuity at date 1, and then, the present value of a lump sum at date 0:

```python
"""
Date                   0               1                2               3              Forever
                       |---------------|----------------|---------------|----------------|
Cash Flow                                          $215,000        215,000(1 + g) 
                       _________________________________|
                       |               |
Present Value     $3,257,575       $3,583,333
"""
```

From the exercise's text we get the following variables:

```python
r = .10 # Discount rate
g = .04 # Growth rate
c_flow = 215000 # first cash flow
```

Lets define a helper function for the growing perpetuity equation:

```python
def growing_perpetuity(c_flow, r, g):
    g_perpetuity = (c_flow / (r - g))
    return g_perpetuity
```

The present value of the growing perpetuty at date 1 is:

```python
perpetuity = growing_perpetuity(c_flow, r, g)
print(f"${perpetuity:,.2f}")
```

So the value of the perpetuity is 3.6 million dollars at date one, but we must find the present value of the perpetuity at date 0:

```python
n = 1
pv = -npf.pv(fv=perpetuity, rate=r, nper=n, pmt=0)
print(f"Present Value at date 0: ${pv:,.2f}")
```

What is the value `today` of a `15-year annuity` that pays `750 a year`?  The annuity's `first payment` occurs `six years from today`.  The annual interest rate is `12 percent` for `years 1 through 5`, and `15 percent thereafter`.

To find the present value of the 15-year annuity today, we need to find the present value of the annuity at date 5 and then find the present value of the lump sum at date 0.

```python
"""
Date              0      1     2     3     4     5     6     7 ... 20
                  |------|-----|-----|-----|-----|-----|-----|-----|
Cash Flow                                             750   750   750
                  _____________________________________|    
                 |                               |
Present Value   2,488                           4,385.53
"""
```

```python
r1 = .12 # Interest rate from year 1 through 5
r2 = .15 # Interest rate thereafter
n = 15 # First 5 periods
pmt = 750 # Per year
```

1. Find the present value of the annuity at date 5:

```python
annuity_pv = -npf.pv(fv=0, rate=r2, nper=n, pmt=pmt)
print(f"Present value at date 5: ${annuity_pv:,.2f}")
```

2. Present value at date 0:

```python
pv = -npf.pv(fv=annuity_pv, rate=r1, nper=5, pmt=0)
print(f"Present value at date 0: ${pv:,.2f}")
```
