## Example 4.1 - Present Value

Linda Jennings, a  financial analys at kaufman & Broad, a leading real estate firm, is thinking about recommending that Kaufman & Broad invest in a piece of land that costs 85,000.  She is certain that next year the land will be worth 91,000, a sure 6,000 gain.  Given that the guaranteed interest rate in the bank is 10 percent, should Kaufman & Broad undertake the investment in land?

A moment's though should be all it takes to convice her that this is not an attractive business deal.  By investing 85,000 in the land, she will have 91,000 available next year.  Suppose instead, that Kaufman & Broad puts the same 85,000 into the bank.  At the interest rate of 10 percent, this 85,000 would grow to:

```python
cost = 85000
fv = (1 + 10) * cost 
print(f' ( 1 + .10 ) x ${cost:,.0f} = ${fv:,.0f} next year.')
```

It would be foolish to buy the land when investing the same 85,000 in the financial market would produce an extra 2,500 (that is, 93,500 from the bank minus 91,000 from the land investment).  This is a future value calculation.

Alternatively, she could calculate the present value of the sale price next year as:

```python
rate = .10
land_fv = 91000
land_pv = land_fv / (1 + rate)
print(f'Present Value = {land_fv:,.0f} / ( 1 + rate ) = {pv:,.0f}')
```

Because the present value of next year's sales price is less than this year's purchase price of 85,000, present value analysis also indicates that she should not recommend purchasing the property.

## Example 4.2 - Uncertainty and Valuation

Profesional Artworks, Inc., is a firm that speculates in modern paintings.  The manager is thinking of buying an original Picasso of 400,000 with the intention of selling it at the end of one year.  The manager expects that the painting will be worth 480,00 in one year. 

Of course, this is only an expectation, the painting could be worth more or less that 480,000.  Suppose the guaranteed interest rate granted by banks is 10 percent.  Should the firm purchase the piece of art?

Our first thought might be to discount at the interest rate, yielding:

```python
rate = .10
art_fv = 480000
art_pv = art_fv / (1 + rate) 
print(f'${art_fv:,.0f} / 1.10 = ${art_pv:,.0f}')
```

Because 436,364 is greater thatn 400,000, it looks at first glance as if the painting should be purchased.  However, 10 percent is the return one can earn on a riskless investment.  Because the painting is quite risky, a higher discount rate is called for.  The manager chooses a rate of 25 percent to reflect this risk.  In other workds, he argues that a 25 percent expected return is fair compensation for an investment as risky as this painting.

The present value of the painting becomes:



```python
rate = .25
art_fv = 480000
art_pv = art_fv / (1 + rate) 
print(f'${art_fv:,.0f} / 1.10 = ${art_pv:,.0f}')
```

Thus, the manager believes that the painting is currently overpriced at 400,000 and does not make the purchase.

## Example 4.5 - Finding the Rate

Carl Voigt, who recently won 10,000 in the lottery, wants to buay a car in five years.  Carl estimates that the car will cost 16,105 at that time.  What interest rate must he earn to be able to afford the car?

The ratio of purchase price to initial cash is:

```python
fv = 16105
pv = 10000
ratio = fv / pv
print(f'${fv:,.0f} / ${pv:,.0f} = {ratio:.4f}')
```

Thus, he must earn an interest rate that allows 1 to become 1.6105 in five years.  We can express the problem algebraically as:

```python
print(f'${pv:,.0f} x (1 + r)^5 = {fv:,.0f}')
```

where r is the interest rate needed to purchase the car.  Because 16,105/10,000 = 1.6105, we have:

```python
r = ratio**(1/5) - 1
print(f'(1 + r)^5 = {ratio:.4f}\nr = {r*100:.0f}%')
```

Either a table, a spreadsheet, or a hand calculator lets us solve for r. (or python)

```python
rate = npf.rate(pv=-pv, fv=fv, nper=5, pmt=0)
print(f'r = {rate*100:.0f}%')
```

## Example 4.7 - Multiperiod Discounting

Bernard Durmas will receive 10,000 three years from now.  Bernard can earn 8 percent on his investments, so the appropriate discount rate is 8 percent.  What is the present value of his future cash flow?  The answer is:

```python
r = .08
n = 3
fv = 10000
pv_fact = 1 / (1 + r)**n
pv = fv * pv_fact
print(f'PV = ${fv:,.0f} x (1 / (1 + {r})^3\nPV = {fv:,.0f} x {pv_fact:.4f}\nPV = {pv:,.0f}')
```

When his investments grow at an 8 percent rate of interest, Bernard Durmas is equally inclined toward receiving 7,938 now and receiving 10,000 in three years' time.  After all, he could convert the 7,938 he receives today into 10,000 in three years by lending it at an interest rate of 8 percent.

## Example 4.10 - Cash Flow Valuation

Kyle Mayer has won the Kentucky State Lottery and will receive the following set of cash flows over the next two years:

1. 20,000
2. 50,000

Mr. Mayer can currently earn 6 percent in his money market account, so the appropriate discount rate is 6 percent.  The prsent value of the cash flows is:

```python
r = .06
cf = [0, 20000, 50000]
pv_of_cash_flows = 0

print("Year", "Cash_Flow", "PV_Factor", "Present_Value", sep=" "*4)

for k, v in enumerate(cf[1:], start=1):
    pv_fact = (1 / (1 + r))**(k)
    pv = v * pv_fact
    pv_of_cash_flows += pv
    print(f"   {k}     {v:,.0f}       {pv_fact:.4f}        {pv:,.2f}")

print(f"Total                               {pv_of_cash_flows:,.2f}")
```

In other words, Mr. Mayer is equally inclined toward receiving 63,367.75 today and receiving 20,000 and 50,000 over the next two years.

```python
# Manual PV calculation shown above for illustration purposes. 
# npf.npv could be used instead.

pv_of_cf = npf.npv(r, cf)
print(f"${pv_of_cf:,.2f}")
```

## Example 4.11 - Net Present Value

Finance.com has an opportunity to invest in a new high-speed computer that costs 50,000.  The computer will generate cash flows (from cost savings) of 25,000 on year from now, 20,000 two years from now, and 15,000 three years from now.  The computer will be worthless after three years, and no additional cash flows will occur.  Finance.com has determined that the appropriate discount rate is 7 percent for this investment.  Should finance.com make this investment in a new high-speed computer?  What is the net present value of the investment?

The cash flows and present valu factors of the proposed computer are s follow:

```python
"""
             Cash_Flow             PV_Factor
Year 0        -50,000                1 = 1
Year 1         25,000           1/1.07 = .9346
Year 2         20,000       (1/1.07)^2 = .8734
Year 3         15,000       (1/1.07)^3 = .8163
"""
```

The present value of the cash flows is:

Cash flows x Present Value Factor = Present Value

```python
r = .07

# Create a basic data structure
struct = {"Years": [0, 1, 2, 3],
          "Cash_Flows": [-50000, 25000, 20000, 15000],
          "PV_Factor": [],
          "Present_Value": []
         }

# Manually discount Cash Flows 
# to get Pv_factors and present values
for t in struct["Years"]:
    if t == 0:
        pv_fact = 1
        pv = struct["Cash_Flows"][t] * pv_fact
    else:
        pv_fact = 1 / (1 + r)**t
        pv = struct["Cash_Flows"][t] * pv_fact

    struct["PV_Factor"].append(pv_fact)
    struct["Present_Value"].append(pv)

# Create table (dataframe)
df = pd.DataFrame(struct)

# Add a total row
total_row = pd.DataFrame({
    "Years": ["Totals"],
    "Cash_Flows": [np.sum(df["Cash_Flows"])],
    "PV_Factor": [np.nan],
    "Present_Value": [np.sum(df["Present_Value"])]
})

# Add total row to dataframe
total_df = pd.concat([df, total_row], ignore_index=True)

# Set Decimals to 2 and 4 places as applicable
total_df["Cash_Flows"] = total_df["Cash_Flows"].map("{:,.2f}".format) 
total_df["PV_Factor"] = total_df["PV_Factor"].map("{:,.4f}".format)
total_df["Present_Value"] = total_df["Present_Value"].map("{:,.2f}".format)

# Display the dataframe
display(HTML(total_df.to_html(index=False)))
```

```python
# Just to make sure
npv = npf.npv(r, struct['Cash_Flows'])
print(f'Net Present Value = {npv:,.2f}')
```

## Example 4.12 - Effective Annual Rates

What is the end-of-year wealth if Jane Christine receives a stated annual interest rate of 24 percent compounded monthly on a 1 dollar investment?

Using Equation 4.6: 

*C0 * (1 + (r / m))^m*  

```python
r = .24
m = 12
investment = 1

wealth = investment * (1 + (r / m))**m
print(f"her wealth is: ${wealth:.4f}")
```

The annual rate of return is 26.82 percent.  This annual rate of return is called either the **effective annual rate (EAR)** or the **effective annual yield (EAY).**  Due to compounding, the effective annual interest rate is grater than the stated annual interest rate of 24 percent. Algebraically, we can rewrite the effective annual interest rate as follows:

**Effective Annual Rate**:

*EAR = (1 + r / m)^m - 1*

## Example 4.19 - Paying Dividends

Popovich corporation is just about to pay a dividend of 3.00 per share.  Investors anticipate that the annual dividend will rise by 6 percent a year forever.  The applicable discount rate is 11 percent.  What is the price of the stock today?

**Equation 4.12**
*PV = C / ( r - g)*

The numerator in Equation 4.12 is the cash flow to be received net period.  Since the growth rate is 6 percent, the dividend next year is:

```python
r = .11
g = .06
current_dividend = 3
future_dividend = current_dividend * (1 + g)
print(f"${future_dividend:.2f}")
```

The price of the stock today is:

```python
pv_of_divs = future_dividend / (r - g)
stock_price = current_dividend + pv_of_divs
print(f"${stock_price:.2f}")
```

The price of 66.60 includes both the dividend to be received immediately and the present value of all dividends beginning a year from now.  Equation 4.12 makes it possible to calculate only the present value of all dividends beginning a year from now.
