**1. Calculating Returns**

Suppose a stock had an initial price of 92 per share, paid a dividend of 1.45 per share during the year, and had an ending share price of 104.  **Compute the Percentage Total Return.**

```python
initial_price = 92
dividend = 1.45
ending_price = 104

Percentage_Total_Return = (dividend / initial_price) + ((ending_price - initial_price) / initial_price)
ptr = f'Percentage Total Return: {Percentage_Total_Return*100:.2f}%'
ptr
```

**2. Calculating Yields**

In Problem 1, what was the dividend yield? The Capital Gains Yield?

```python
dividend_yield = (dividend / initial_price)
capital_gains_yield = ((ending_price - initial_price) / initial_price)

div_yield = f'Dividend Yield: {dividend_yield*100:.2f}%'  
cap_yield = f'Capital Gains Yield: {capital_gains_yield*100:.2f}%'

print(f'{div_yield}\n{cap_yield}')
```

**4. Calculating Returns**

Suppose you bought an 8 percent coupon bond one year ago for 1,090.  The bond sells for 1,056 today.

1. Assuming a 1,000 face value, what was your toal dollar return on this investment over the past year?
2. What was your total nominal rate of return on this investment over the past year
3. If the inflation rate last year was 3 percent, what was your total real rate of return on this investment?

```python
beginning_price = 1090
ending_price = 1056
coupon = .08

# 1. Total Dollar Return
face = 1000
total_dollar_return = (ending_price - beginning_price) + (face * coupon)
print(f'Total Dollar Return on Investment: ${total_dollar_return:.2f}')

# 2. Nominal Rate of Return
nominal_rate_of_return = ((ending_price - beginning_price) + (face * coupon)) / beginning_price
print(f'Nominal Rate of Return: {nominal_rate_of_return*100:.2f}%')

# 3. Total Real Rate of Return
inflation = .03
real_rate = (1 + nominal_rate_of_return)/(1 + inflation) - 1 # Fisher effect
print(f'Real Rate of Return: {real_rate*100:.2f}%')
```

**5. Nominal versus Real Returns**

What was the arithmetic average annual return on large-company stocks from 1926 through 2008:

1. In nominal terms?
2. In real terms? 

```python
inflation = total_returns["Consumer Price Index"].mean() / 100
nominal_terms = total_returns["Large-Company Stock"].mean() / 100
real_terms = (1 + nominal_terms)/(1 + inflation) - 1 # Fisher effect
print(f"Nominal terms: {nominal_terms*100:.2f}%\nReal terms: {real_terms*100:.2f}%")
```

**6. Bond Returns**

What is the historical real return on: 
1. Long-term government bonds?
2. Long-term corporate bonds?

```python
# inflation
inflation = total_returns["Consumer Price Index"].mean() / 100

# Long-term government bonds
government_bonds = total_returns["Long-Term Government Bonds"].mean() / 100
real_government_bonds = (1 + government_bonds) / (1 + inflation) - 1 # Fisher effect

# Long-term corporate bonds
# Rate of 6.2 obtained from Table 10.3, Page 317, Corporate Finance 9th
corporate_bonds = 6.2 / 100 
real_corporate_bonds = (1 + corporate_bonds) / (1 + inflation) - 1 # Fisher effect

print(f"Long-term Government Bonds: {real_government_bonds*100:.2f}")
print(f"Long-term Corporate Bonds: {real_corporate_bonds*100:.2f}")
```

**7. Calculating Returns & Variability** 

Using the following returns, calculate the average returns, the variances, and the standard deviations for X and Y:

```python
"""
Year      X       Y
----     ---     ---
 1       15%     18%
 2       23      29
 3      -34     -31
 4       16      19
 5        9      11
----     ---    ---
"""
```

```python
# Easier if we just group our data in a dataframe
df = pd.DataFrame([[1, .15, .18],
                   [2, .23, .29],
                   [3, -.34, -.31],
                   [4, .16, .19],
                   [5, .09, .11]],
                 columns=("Year", "X", "Y")
)

average_row = pd.DataFrame({
    "Year": ["Mean"],
    "X": [df["X"].mean()],
    "Y": [df["Y"].mean()]
})

variance_row = pd.DataFrame({
    "Year": ["Variance"],
    "X": [np.var(df["X"], ddof=1)],
    "Y": [np.var(df["Y"], ddof=1)]
})

stdev_row = pd.DataFrame({
    "Year": ["StDev"],
    "X": [np.std(df["X"], ddof=1)],
    "Y": [np.std(df["Y"], ddof=1)]
})

df = pd.concat([df, average_row, variance_row, stdev_row], ignore_index=True)
display(HTML(df.to_html(index=False)))
```
