**4. Calculating Returns**

Suppose you bought an 8 percent coupon bond one year ago for 1,090.  The bond sells for 1,056 today.

1. Assuming a 1,000 face value, what was your toal dollar return on this investment over the past year?
2. What was your total nominal rate of return on this investment over the past year
3. If the inflation rate last year was 3 percent, what was your total real rate of return on this investment?

```python
beginning_price = 1090
ending_price = 1056
coupon = .08

# 1. Total Dollar Return
face = 1000
total_dollar_return = (ending_price - beginning_price) + (face * coupon)
print(f'Total Dollar Return on Investment: ${total_dollar_return:.2f}')

# 2. Nominal Rate of Return
nominal_rate_of_return = ((ending_price - beginning_price) + (face * coupon)) / beginning_price
print(f'Nominal Rate of Return: {nominal_rate_of_return*100:.2f}%')

# 3. Total Real Rate of Return
inflation = .03
real_rate = (1 + nominal_rate_of_return)/(1 + inflation) - 1 # Fisher effect
print(f'Real Rate of Return: {real_rate*100:.2f}%')
```

**7. Calculating Returns & Variability** 

Using the following returns, calculate the average returns, the variances, and the standard deviations for X and Y:

```python
"""
Year      X       Y
----     ---     ---
 1       15%     18%
 2       23      29
 3      -34     -31
 4       16      19
 5        9      11
----     ---    ---
"""
```

```python
# Easier if we just group our data in a dataframe
df = pd.DataFrame([[1, .15, .18],
                   [2, .23, .29],
                   [3, -.34, -.31],
                   [4, .16, .19],
                   [5, .09, .11]],
                 columns=("Year", "X", "Y")
)

average_row = pd.DataFrame({
    "Year": ["Mean"],
    "X": [df["X"].mean()],
    "Y": [df["Y"].mean()]
})

variance_row = pd.DataFrame({
    "Year": ["Variance"],
    "X": [np.var(df["X"], ddof=1)],
    "Y": [np.var(df["Y"], ddof=1)]
})

stdev_row = pd.DataFrame({
    "Year": ["StDev"],
    "X": [np.std(df["X"], ddof=1)],
    "Y": [np.std(df["Y"], ddof=1)]
})

df = pd.concat([df, average_row, variance_row, stdev_row], ignore_index=True)

# Convert returns to percentages
df['X'] = df['X'].apply(lambda x: x * 100 if isinstance(x, (int, float)) else x)
df['Y'] = df['Y'].apply(lambda x: x * 100 if isinstance(x, (int, float)) else x)

# Format the values to two decimal places
df['X'] = df['X'].apply(lambda x: f"{x:.2f}%" if isinstance(x, (int, float)) else x)
df['Y'] = df['Y'].apply(lambda x: f"{x:.2f}%" if isinstance(x, (int, float)) else x)

display(HTML(df.to_html(index=False)))
```

**8. Risk Premium**

Refer to Table 10.1 (total_returns) in the text and look athe period from 1973 through 1978.
1. Calculate the arithmetic average returns for large-company stocks and T-Bills over the period.
2. Calculate the standardard deviation of the retuns for large-company stocks and T-Bills over the period.
3. Calculate the observed risk premium in each year for the large-company stocks versus T-Bills.  What was the the arithmetic average risk premium over this period? What was the standard deviation of the risk premium over this period?



```python
# We can answer all this question in one dataframe

start_year = 1973
end_year = 1978

variables = ["Year", "Large-Company Stock", "U.S. Treasury Bills"]
sample = total_returns[variables]
sample = sample[(sample["Year"] >= start_year) & (sample["Year"] <= end_year)         ]
sample["Risk Premium"] = sample["Large-Company Stock"] - sample["U.S. Treasury Bills"]

average_row = pd.DataFrame({
    "Year": ["Mean"],
    "Large-Company Stock": [sample["Large-Company Stock"].mean()],
    "U.S. Treasury Bills": [sample["U.S. Treasury Bills"].mean()],
    "Risk Premium": [sample["Risk Premium"].mean()]
})

stdev_row = pd.DataFrame({
    "Year": ["StDev"],
    "Large-Company Stock": [np.std(sample["Large-Company Stock"], ddof=1)],
    "U.S. Treasury Bills": [np.std(sample["U.S. Treasury Bills"], ddof=1)],
    "Risk Premium": [np.std(sample["Risk Premium"], ddof=1)],
})

sample = pd.concat([sample, average_row, stdev_row], ignore_index=True)

# Format the values to two decimal places
sample['Large-Company Stock'] = sample['Large-Company Stock'].apply(lambda x: f"{x:.2f}%" if isinstance(x, (int, float)) else x)
sample['U.S. Treasury Bills'] = sample['U.S. Treasury Bills'].apply(lambda x: f"{x:.2f}%" if isinstance(x, (int, float)) else x)
sample['Risk Premium'] = sample['Risk Premium'].apply(lambda x: f"{x:.2f}%" if isinstance(x, (int, float)) else x)

display(HTML(sample.to_html(index=False)))
```

**9. Calculating Returns and Variability**

You've observed the following returns on Mary Ann Data Corporation's stock over the past five years:

1.  34%
2.  16%
3.  19%
4. -21%
5.   8%

Questions:
1. What was the arithmetic average return on Mary Ann's stock over the five-year period?
2. What was the variance of Mary Ann's returns over this period? The Standard Deviation?
   

```python
returns = pd.DataFrame([["1", .34],
                        ["2", .16], 
                        ["3", .19], 
                        ["4", -.21], 
                        ["5", .08]],
                      columns=("Year", "Returns"))

average_row = pd.DataFrame({
    "Year": ["Mean"],
    "Returns": [returns["Returns"].mean()]
})

variance_row = pd.DataFrame({
    "Year": ["Variance"],
    "Returns": [np.var(returns["Returns"], ddof=1)]
})

stdev_row = pd.DataFrame({
    "Year": ["StDev"],
    "Returns": [np.std(returns["Returns"], ddof=1)]
})

returns = pd.concat([returns, average_row, variance_row, stdev_row], ignore_index=True)

# Convert returns to percentages
returns['Returns'] = returns['Returns'].apply(lambda x: x * 100 if isinstance(x, (int, float)) else x)

# Format the values to two decimal places
returns['Returns'] = returns['Returns'].apply(lambda x: f"{x:.2f}%" if isinstance(x, (int, float)) else x)

display(HTML(returns.to_html(index=False)))
```

**10. Calculating Real Returns and Risk Premiums**

In Problem 9, suppose the average inflation rate over this period was 4.2 percent, and the average T-bill rate over the period was 5.1 percent:

1. What was the the average real return on Mary Ann's stock?
2. What was the average nominal risk premium on Mary Ann's stock?

```python
# 1 - Average Real Return
inflation = .042
avg_return = .1120 # From Problem 9.
avg_real_return = (1 + avg_return) / (1 + inflation) - 1 # Fisher effect
print(f"Average Real Return: {avg_real_return*100:.2f}%")

# 2 - Average Nominal Risk Premium
average_tbill = .051
avg_nominal_risk_premium = avg_return - average_tbill
print(f"Average Nominal Risk Premium: {avg_nominal_risk_premium*100:.2f}%")
```

**11. Calculating Real Rates**

Given the information Problem 10: 
1. What was the average real risk-free rate over over this time period?
2. What was the average real risk premium? 

```python
# 1 - Average Real  Risk-Free Premium
avg_real_risk_free_premium = (1 + average_tbill) / (1 + inflation) - 1 # Fisher Effect
print(f"Average Real Risk-Free Rate: {avg_real_risk_free_premium*100:.2f}%")

# 2 - Average Real Risk premium
avg_real_risk_premium = avg_nominal_risk_premium - avg_real_risk_free_premium
print(f"Average Real Risk Premium: {avg_real_risk_premium*100:.2f}%")
```
